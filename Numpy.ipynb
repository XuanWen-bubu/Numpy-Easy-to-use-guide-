```python
import numpy as np

array = np.array([[1, 2, 3], [4, 5, 6]], dtype="int64")
print(array)
print(array.shape)  # 返回元组表示array的维度
print(array.ndim)  # 返回数字表示array维度的数目
print(array.size)  # 返回数字表示array中所有数据元素的数目
```

```python
a = np.array([[1, 2], [4, 5]])
b = np.arange(4).reshape((2, 2))
c = a*b  # 对应元素直接相乘
d_dot = np.dot(a, b)  # 矩阵乘法
d_dot_2 = a.dot(b)  # 矩阵乘法另一种形式
print(c)
print(d_dot)
print(d_dot_2)
```

```python
a = np.array([[0, 2, 3, 4, 5], [6, 7, 8, 9, 10]])
print(np.sum(a, axis=1))
print(np.diff(a))  # 后一位元素减去前一位元素。因为一行5个数，只有前4个数有后一位元素，所以是2行4列
print(np.nonzero(a))  # 第一个array表示行，第二个array表示列，会返回所有非0元素的索引
```

```python
a = np.array([[0, 2, 3, 4, 5], [6, 7, 8, 9, 10]])
print(a)
print(a.T)  # 矩阵转置
print(a.clip(5, 9))  # 比5小的都变为5，比9大的都变为9，5和9之间的不变
```

```python
a = np.arange(36).reshape(9, 4)
print(a)
print(a[[0, 1], [2, 3]])  # 0行2列 1行3列
b = np.array([[0, 1], [2, 3]])
c = np.array([0, 1, 2, 3])
print(f"二维数组用二维数组进行索引结果：\n{a[b]}")  # 二维数组用二维数组进行索引，返回三维数组。0行 1行  2行 3行
print(f"二维数组用一维数组进行索引结果：\n{a[c]}")
print(
    a.T[b]
)  # 将同样一个二维数组转置后用二维数组进行索引，返回三维数组。返回的是原二维数组的0列 1列  2列 3列
```

```python
a = np.arange(36).reshape(9, 4)
print(a)
print(a.flatten())  # 将一个多维数组转为一维数组，返回一个一维数组
print(a.flat)
for item in a.flat:  # 返回一个迭代器
    print(item)
```

```python
# array合并
a = np.array([1, 1, 1])
b = np.array([2, 2, 2])
c = np.vstack((a, b, a))  # vertical stack:竖直方向合并
d = np.hstack((a, b))  # horizontal stack:水平方向合并
print(f"竖直合并结果:\n{c}")
print(f"水平合并结果:\n{d}")
e = a[:, np.newaxis]
f = b[:, np.newaxis]
print(f"转置后竖直合并结果:\n{np.vstack((e, f))}")
print(f"转置后水平合并结果:\n{np.hstack((e, f))}")
g = np.concatenate((e, f), axis=0)  # concatenate:可以用axis参数选择合并的方向
h = np.concatenate((e, f), axis=1)
print(g)
print(h)
```

```python
# 一维数组进行转置的方法
# ①reshape(-1,1)
a = np.array([1, 1, 1])
# numpy会根据数组的总元素数量以及其他指定维度的大小,来确定-1对应的维度大小。
# (-1,1)表示将所有元素排成一列,行数根据数组的总元素数量自动计算得出
print(f"一维数组转置:\n{a.reshape(-1, 1)}")
a = np.array([1, 1, 1, 2, 2, 2])
print(f"一维数组转为两列:\n{a.reshape(-1, 2)}")
# ②mat:接收多种输入（元组，列表，数组），将其转化为矩阵对象
b = np.mat(a)
print(b.T)
# ③newaxis:用于新增一个维度
# a.shape是(6,),a[:,np.newaxis].shape是(6, 1),a[np.newaxis,:].shape是(1,6)
# a[:,np.newaxis]在第1个位置插入1个维度，(6,)变为(6,1);a[np.newaxis,:]在第0个位置插入1个维度，(6,)变为(1,6)
print(a[:, np.newaxis], a[:, np.newaxis].shape)
print(a[np.newaxis, :], a[np.newaxis, :].shape)
```

```python
# array分割
a = np.arange(12).reshape(3, 4)
print(a)
print(np.split(a, 2, axis=1))  # split(数组,分割的份数,分割的方向)
# print(np.split(a,2,axis=0)) split必须进行等比例分割，如果不等比例会报错
print(np.array_split(a, 2, axis=0)) # 不等比例的分割可以用array_split
print(np.vsplit(a,3)) # 竖直方向分割
print(np.hsplit(a,4)) # 水平方向分割
```

```python
a = np.arange(36).reshape((6, 6))
print(a)
print(a[[2, 2, 4, 4], [1, 3, 1, 3]])
# [[2, 2], [4, 4]]是行索引数组，[[1, 3], [1, 3]]是列索引数组，按位置对应提取元素
print(a[[[2, 2], [4, 4]], [[1, 3], [1, 3]]])
```
